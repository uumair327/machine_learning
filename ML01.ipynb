```python
# Assuming 'price' is the target variable and 'area' is the predictor
X = df['area'].values
Y = df['price'].values

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 20)
# hyper parameter "random state"
```

Linear Regression From Scratch Without any Library using loss function

```python
import pandas as pd
data = pd.read_csv("/content/sample_data/BankChurners.csv")
print(data.head())
```

```python
X = data['Customer_Age']; Y = data['Months_on_book']
X = X.tolist()
Y = Y.tolist()
```

```python
#alpha - learning rate
def gradient_descent(X, Y, w, b, alpha):

    dl_dw = 0.0
    dl_db = 0.0
    N = len(X)

    for i in range(N):
        dl_dw += -1*X[i] * (Y[i] - (w*X[i] + b))
        dl_db += -1*(Y[i] - (w*X[i] + b))

    w = w - (1/float(N)) * dl_dw * alpha
    b = b - (1/float(N)) * dl_db * alpha

    return w, b
```

```python
def cost_function (X, Y, w, b):

    N = len(X)
    total_error = 0.0
    for i in range(N):
        total_error += (Y[i] - (w*X[i] - b))**2 # mean Sq error

    return total_error / (2*float(N))
```

```python
def train(X, Y, w, b, alpha, n_iter):

    for i in range(n_iter):
        w, b = gradient_descent(X, Y, w, b, alpha)

        if i % 400 == 0:
            print ("iteration:", i, "cost: ", cost_function(X, Y, w, b))


    return w, b
```

```python
def predict(x, w, b):
	return x*w + b
```

```python
w, b = train(X, Y, 0.0, 0.0, 0.0001, 7000)
x_new = 50.0
y_new = predict(x_new, w, b)
print(y_new)
```

```python
import matplotlib.pyplot as plt
# Predict values for plotting
X_range = range(int(min(X)), int(max(X)) + 1)  # range of X values
Y_pred = [predict(x, w, b) for x in X_range]

# Plot data points and regression line
plt.scatter(X, Y, s=5, label='Data points')
plt.plot(X_range, Y_pred, color='red', linewidth=2, label='Regression line')
plt.grid()
plt.xlabel("Customer_Age values")
plt.ylabel("Months_on_book values")
plt.legend()
plt.show()
```
